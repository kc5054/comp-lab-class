```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import MDAnalysis as mda
from MDAnalysis.analysis import rdf
from MDAnalysis.transformations import boxdimensions
```

```python
energyData=np.loadtxt('proton_aimd_1ns_every0.5fs-1.ener',skiprows=1)
```

```python
kin=energyData[:,2]
pot=energyData[:,4]
time=energyData[:,1]
time=np.asarray(time)
time/=1000
kin=np.asarray(kin)
pot=np.asarray(pot)
```

```python
fig,ax=plt.subplots()
ax.plot(time,kin+pot)
ax.set_title("Energy - time (proton)")
ax.set_xlabel('Time (ps)')
ax.set_ylabel('Energy (a.u.)')
```

```python
#traj=md.load('water_aimd_1ns_every0.5fs-pos-1.xyz',top='water.pdb')
traj=mda.Universe('proton.pdb','proton_aimd_1ns_every0.5fs-pos-1.xyz')
dim = [12.42, 12.42, 12.42, 90, 90, 90]
tran=boxdimensions.set_dimensions(dim)
traj.trajectory.add_transformations(tran)
```

```python
ag=traj.select_atoms('name O')
```

```python
gr=rdf.InterRDF(ag,ag,nbins=60,range=(0,6.0))
gr.run()
```

```python
fig,ax=plt.subplots()
ax.plot(gr.bins,gr.rdf)
ax.set_xbound(1,6)
ax.set_ybound(0,5)
ax.set_xlabel('r (Angstrom)')
ax.set_ylabel('gr')
ax.set_title('RDF (proton)')
```
