```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('density.xvg',names=['Time','density'],skiprows=24,sep='\s+')
data['Time']/=1000
fig,ax=plt.subplots()
ax.set_xlabel('Time(ns)')
ax.set_ylabel('Density')
ax.set_title('Density-Time')
ax.plot(data['Time'],data['density'])
```
