```python
import numpy as np
import mdtraj as md
import matplotlib.pyplot as plt
```

```python
trj=md.load('step5_prod.xtc',top='step5_nowater.gro')
```

```python
pair = np.array([[132, 277]], dtype=np.int32)
dis=md.compute_distances(trj, pair, periodic=False, opt=True)
time=np.zeros((90,1))
for i in range (90):
    time[i]=i
rg=md.compute_rg(trj)
dis2=np.zeros((90,1))
rg2=np.zeros((90,1))
for i in range(90):
    dis2[i]=dis[i*10]
    rg2[i]=rg[i*10]
fig, ax1 = plt.subplots()
ax2=ax1.twinx()
ax1.plot(time, dis2, 'g-')
ax2.plot(time,rg2, 'b-')
ax1.set_xlabel('Time (ns)')
ax1.set_ylabel('end_end distance', color='g')
ax2.set_ylabel('Radius of gyration', color='b')
plt.show()
```
