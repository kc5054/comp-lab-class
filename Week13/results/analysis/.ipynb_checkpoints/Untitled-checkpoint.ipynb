```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rcData=np.loadtxt('/home/kc5054/CHEM2671/comp-lab-class/Week13/Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog')
stdrcData=np.loadtxt('/home/kc5054/CHEM2671/comp-lab-class/Week13/Outputs/monitor/MONITOR-COLVAR.metadynLog',skiprows=189)
freqData=np.loadtxt('/home/kc5054/CHEM2671/comp-lab-class/Week13/Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog')
```

```python
time=rcData[:,0]
rc=rcData[:,1]
time=np.asarray(time)
stdtime=stdrcData[:,0]
stdtime=np.asarray(stdtime)
stdrc=stdrcData[:,1]
freqtime=freqData[:,0]
freqtime=np.asarray(freqtime)
freq=freqData[:,1]
```

```python
fig,ax=plt.subplots()
ax.plot(time/1000,rc,label='Height = 1.5, Width = 0.5, Freq = 100')
ax.plot(stdtime/1000,stdrc,label = "Standard")
ax.scatter(freqtime/1000,freq,label='Freq Hill',s=1,color='r')
ax.set_xlabel('time (ns)')
ax.set_ylabel('Reaction coordinates')
ax.set_title("Reaction Coordinates - Time")
plt.legend()
```
