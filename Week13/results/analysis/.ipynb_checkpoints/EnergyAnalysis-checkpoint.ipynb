```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nptPotData=np.loadtxt('/home/kc5054/CHEM2671/comp-lab-class/Week13/Outputs/NPT-1.ener',skiprows=1)
nvtPotData=np.loadtxt('/home/kc5054/CHEM2671/comp-lab-class/Week13/Outputs/NVT-1.ener',skiprows=1)
qmmmPotData=np.loadtxt('/home/kc5054/CHEM2671/comp-lab-class/Week13/Outputs/monitor/MONITOR-1.ener',skiprows=1)
```

```python
nptPot=nptPotData[:,4]
nvtPot=nvtPotData[:,4]
timeNPT=nptPotData[:,1]
timeNVT=nvtPotData[:,1]
timeNPT=np.asarray(timeNPT)
timeNVT=np.asarray(timeNVT)
timeQM=qmmmPotData[:,1]
qmPot=qmmmPotData[:,4]
timeQM=np.asarray(timeQM)
```

```python
fig,ax=plt.subplots()
ax.plot(timeNPT/1000,nptPot,label='NPT')
ax.set_xlabel('Time (ps)')
ax.set_ylabel('Potential Energy [a.u.]')
ax.set_title('Potential Energy - Time (NPT)')
```

```python
fig,ax=plt.subplots()
ax.plot(timeNVT/1000,nvtPot,label='NVT')
ax.set_xlabel('Time (ps)')
ax.set_ylabel('Potential Energy [a.u.]')
ax.set_title('Potential Energy - Time (NVT)')
```

```python
fig,ax=plt.subplots()
ax.plot(timeQM/1000,qmPot,label='QM/MM')
ax.set_xlabel('Time (ps)')
ax.set_ylabel('Potential Energy [a.u.]')
ax.set_title('Potential Energy - Time (QM/MM)')
```
