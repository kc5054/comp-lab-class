```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('1hz3_T310.run.25000000.energy.xvg',header=0,names=['Time(ns)','Potential','KineticEnergy','TotalEnergy','Temperature','Pressure','Volume'],skiprows=29,sep='\s+')
data['Time(ns)']/=1000
data.plot(x='Time(ns)',y='Potential')
```

```python
data.plot(x='Time(ns)',y="KineticEnergy")
```

```python
data.plot(x='Time(ns)',y="TotalEnergy")
```

```python
data.plot(x='Time(ns)',y='Pressure')
```

```python
data.plot(x='Time(ns)',y='Volume')
```

```python
data.plot(x='Time(ns)',y='Temperature')
```

```python
fig, ax1 = plt.subplots()
ax2=ax1.twinx()
m=data['Time(ns)']
n=data['Temperature']
l=data['KineticEnergy']
ax1.plot(m, n, 'g-')
ax1.set_yticks(np.arange(302.5,330,2.5))
ax2.plot(m,l, 'b-')
ax1.set_xlabel('Time (ns)')
ax1.set_ylabel('Temperature', color='g')
ax2.set_ylabel('KE', color='b')
plt.show()
```
